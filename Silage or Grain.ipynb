```python
# Renaming Sowing Date Columns

def rename_sowing_date(col_val):

    index_to = col_val.find("to")
    index_to = index_to + 3

    col_val = col_val[index_to:]
    
    col_val = col_val.replace(' ','-')
    col_val = col_val.replace('th','')

    return col_val

sowing_date_og = df['Sowing Date']

df['Sowing Date'] = df['Sowing Date'].apply(rename_sowing_date)
df


```

```python
Name = []
arr = []

row_sowing_date = list(df['Sowing Date'])





for row_index in range(len(df)):

    row = df.iloc[row_index].tolist()

    row = row[1:-2]
    
    ans = "Grain"
   
    diff = 0
    
    start = cols_dict[row_sowing_date[row_index]] + 1
    arr.append(start)
    
    
    for i in range(start,len(row)):

        if i != (len(row) - 1):

            next_ = row[i + 1]
            prev_ = row[i]

            diff = 1 - (next_ / prev_)

            if diff >= 0.5:

                if i != (len(row) - 2):

                    if row[i + 2] <= next_:

                        ans = "Silage"
                        break
                    else:
                        continue
                else:
                    ans = "Silage"
                    break

            elif diff <= 0.3:
                ans = "Grain"

            elif diff > 0.3 and diff <= 0.4:
                ans = "Blank-1"
                break
            elif diff > 0.4 and diff < 0.5:
                ans = "Blank-2"
                break
            

    Name.append(ans)
```
